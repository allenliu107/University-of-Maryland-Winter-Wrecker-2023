```python
import pandas as pd
```

```python
#8.2 Reading CSV Data into a DataFrame
df = pd.read_csv(r"/Users/allen/Documents/Python/INST 126/Final Project/\Users\allen\Documents\Python\INST 126\Final Project\WinterWrecker.csv")
df
```

```python


# Add a new column 'pound weight' calculated from 'weight' column which is the weight in kilos
#8.1 Used an narray or Series to perform a vectorized computation instead of a loop or comprehension.
df.loc[:, 'pound weight'] = df['Weight'] * 2.2

# Move 'pound weight' column to be the 7th column
column_order = list(df.columns)
column_order.insert(6, column_order.pop(-1))  
# Move last element to the 6th position

# Reorder the DataFrame columns
df = df[column_order]

print(df)
```

```python
#Finding all of the Junior Lifters
```

```python
# Remove the tilde (~) symbol for strings and keep integers as they are
df['Age'] = df['Age'].apply(lambda x: x.replace('~', '') if isinstance(x, str) else x)

# Convert the 'Age' column to numeric, handling errors by setting non-convertible values to NaN
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')

# Drop rows with NaN values (optional, depending on your requirements)
df = df.dropna()

# Perform the comparison after conversion
#8.3 Using Pandas to get a subset of a data frame using boolean
subset_df = df[df['Age'] <= 22]

print(subset_df)
```

```python
#Finding all of UMD lifters
```

```python
#names to check
names_to_check = ['Guodong Liu', 'Guoliang Liu', 'Kyle Pak', 'Jorge Zheng', 'Eli DeLuca', 'Isaiah Lee', 'Anmol Singh', 'Noah Pak', 'Shannon Van', 'Joshua Pak', 'Qiuyu Xue', 'Ariana Boyd']

# Create a boolean mask for the subset of names you want to check
mask = df['Lifter'].isin(names_to_check)

# Use the boolean mask to filter the DataFrame
#8.3 Using Pandas to get a subset of a data frame using boolean
subset_df = df[mask]

# Display the subset DataFrame
print(subset_df)
```
